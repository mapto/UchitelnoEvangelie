```python
for d in docs:
    d.clean_hyphens()
```

```python
html_template = ""
with open("template.html", "r") as ftempl:
    html_template = ftempl.read()

html_row_template ="<tr><td class='index'>%s</td><td>%s</td></tr>"
html_content = html_row_template % ("", "</td><td>".join(["." + d.ext.upper() for d in docs]))
for page in docs[0].pages().keys():
    mpage = [d.pages()[page] for d in docs] 
    mlines = max([len(p) for p in mpage])
    for l in range(mlines):
        mline = [format[l] if len(format) > l else "" for format in mpage]
        line = "%s%d" % (page, l + 1)
        html_content += html_row_template % (line, "</td><td>".join(mline))
# print(html_content)
IPython.display.HTML(html_template % html_content)
# IPython.display.Markdown(html_template % html_content)
# IPython.display.IFrame(src='data:text/html;charset=utf-8,' + (html_template % html_content), width=900, height=600, allow="")
```

```python
tex_template = ""
with open("template.tex", "r") as ftempl:
    tex_template = ftempl.read()

tex_row_template ="%s  &  %s"
tex_content = tex_row_template % ("", "  &  ".join(["." + d.ext.upper() for d in docs]))
for page in docs[0].pages().keys():
    mpage = [d.pages()[page] for d in docs] 
    min_lines = min([len(p) for p in mpage])
    for l in range(min_lines):
        mline = [format[l] for format in mpage]
        line = "%s%02d" % (page, l)
        tex_content += tex_row_template % (line, "  &  ".join(mline)) + " \\"
# print(html_content)
IPython.display.Latex(tex_template % tex_content)
```
